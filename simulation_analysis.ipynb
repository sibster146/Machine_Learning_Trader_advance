```python

import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
simulation = pd.read_csv("simulations/simulation_4_3_update_lag_xgbclassifier2.pkl.csv")
plt.figure(figsize=(10, 5))
plt.plot(simulation["sell_sequence_number"], simulation["pnl"], label="PNL", color='blue', linewidth=2)

plt.xlabel("Sequence Number")
plt.ylabel("Change in Value ($)")
plt.title("PNL")
plt.legend()
plt.grid(True)
plt.show()
```
