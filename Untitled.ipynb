```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', None)
```

```python
def graph(y_actual, y_pred, buy_prices, pnl, backlog_pred, start = None, end = None, tick = 0):

    if not start:
        start = 0
    if not end:
        end = len(y_actual)

    
    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(len(buy_prices))[start:end], buy_prices[start:end], label="Buy", color='green', linewidth=2)
    plt.plot(np.arange(len(y_actual))[start:end], y_actual[start:end], label="Actual", color='blue', linewidth=2)
    plt.plot(np.arange(len(y_pred))[start:end], y_pred[start:end], label="Predicted", color='red', linestyle='dashed', linewidth=2)
    #plt.plot(np.arange(len(backlog_pred))[start:end], backlog_pred[start:end], label="Backlog", color='orange', linestyle='dashed', linewidth=2)

    plt.xlabel("Time Steps")
    plt.ylabel("Value")
    if tick:
        plt.xticks(np.arange(start, end, tick))
    plt.title("Actual vs Predicted Over Time")
    plt.legend()
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(len(pnl))[start:end], pnl[start:end], label="PNL", color='blue', linewidth=2)
    plt.xlabel("Time Steps")
    plt.ylabel("Value")
    plt.title("PnL")
    
    plt.legend()
    plt.grid(True)
    plt.show()
    
```

```python
df = pd.read_csv("simple_simulations/BTC-USD_2025-03-04 16:41:37.753667_50_new_update_linear_regressor.pkl.csv")
mid_df = df[df["type"]==0]
mid_seq_num = mid_df["sequence_number"]
mid_price = mid_df["price"]

pred_df = df[df["type"]==1]
pred_seq_num = pred_df["sequence_number"]
pred_price = pred_df["price"]

plt.figure(figsize=(10, 5))
plt.plot(mid_seq_num[:500], mid_price[:500], label="Mid Price", color='green', linewidth=2)
plt.plot(pred_seq_num[:500], pred_price[:500], label="Pred Price", color='yellow', linewidth=2)
plt.xlabel("Sequence Number")
plt.ylabel("Value")
plt.title("Actual vs Predicted Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
